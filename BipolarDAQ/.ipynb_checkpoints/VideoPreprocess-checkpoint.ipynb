```python
import moviepy.editor as mp
import matplotlib.pyplot as plt
import scipy.io.wavfile as sciwav
import scipy.signal as sps
import numpy as np
from IPython.display import Audio
```

```python
'''
General Functions
'''

def aPlot(figName, is3D = False):
    ax = []
    
    fig1 = plt.figure(figsize = (14,4))
    fig1.suptitle(figName, fontsize=16)
    if(is3D):
        ax = fig1.add_subplot(111, projection='3d')
    else:
        ax = fig1.add_subplot(111)
        
    return ax
```

```python
# opened_clip = mp.AudioFileClip("H:/HapticScreenVideoLibrary/CarChasing2.mp4")
# opened_clip.write_audiofile("./Audio/CarChasing.wav") # NOT working properly!
```

```python
fileName = "CarChasing2"

originalSound = sciwav.read("./Audio/%s.wav" % fileName)

originalFs = originalSound[0]
originalSound = originalSound[1]
sampleNum, channelNum = originalSound.shape

print("Audio sample number = %d [with %d Channel], sampling frequency = %d Hz" % (sampleNum, channelNum, originalFs))
```

```python
t = np.arange(sampleNum)/originalFs

aPlot("Left Channel", is3D = False)
plt.plot(t, originalSound[:,0])
aPlot("Right Channel", is3D = False)
plt.plot(t, originalSound[:,1])
```

```python
cutInd = 14*originalFs
cutSound = originalSound[:cutInd,0]
```

```python
Audio(cutSound, rate=originalFs)
```

```python
Fs = 1000
resampleNum = 14*Fs
resampleSound = sps.resample(cutSound, resampleNum)

aPlot("Left Channel", is3D = False)
plt.plot(t, originalSound[:,0])

t2 = np.arange(resampleNum)/Fs
plt.plot(t2, resampleSound)

plt.xlim(6, 6.2)
```
